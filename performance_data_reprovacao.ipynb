```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""works for year > 2010"""

def create_filtered_dataset_fail_2011_2020(year):
    df = pd.read_excel(f'data/tx_rend_brasil_regioes_ufs_{year}.xlsx', skiprows=5)
    
    #selecting columns
    df = df.iloc[:,:40]
    
    #naming columns
    df.columns = ['Ano', 'Unidade Geográfica', 'Localização',
                  'Rede', 'Total - EF', 'Anos Iniciais - EF',
                  'Anos Finais - EF', '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Total - EM', '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 
                  'R_EF_Total', 'R_EF_Anos Iniciais', 'R_EF_Anos Finais', 'R1EF', 'R2EF', 
                  'R3EF', '4REF', 'R5EF', 'R6EF', 'R7EF','R8EF', 'R9EF', 
                  'RTotal - EM', 'R1EM', 'R2EM', 'R3EM', 'R4EM', 'RNão-Seriado',]
    
    df = df.iloc[3:,:]
    df = df.loc[:20,:]
    df.drop(columns=['Unidade Geográfica', 'Total - EF', 'Anos Iniciais - EF',
                  'Anos Finais - EF', '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Total - EM', '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 
                  'R_EF_Anos Iniciais', 'R_EF_Anos Finais',], inplace=True)
    
    return df
    
```

```python
"""works for year <= 2010"""

def filtered_dataset_fail_from_2007_2010(year, skiprows=6):
    df = pd.read_excel(f'data/tx_rend_brasil_regioes_ufs_{year}.xlsx', skiprows=skiprows)
    
    #selecting columns
    df = df.iloc[:,:40]
    
    #naming columns
    df.columns = ['Ano', 'Unidade Geográfica', 'Rede',
                  'Localização', '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Anos Iniciais - EF', 'Anos Finais - EF', 'Total - EF',
                  '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 'Total - EM', 
                  'R1EF', 'R2EF','R3EF', '4REF', 'R5EF', 'R6EF', 'R7EF','R8EF', 'R9EF', 
                  'R_EF_Anos Iniciais', 'R_EF_Anos Finais', 'R_EF_Total',
                  'R1EM', 'R2EM', 'R3EM', 'R4EM', 'RNão-Seriado', 'RTotal - EM',
                  ]
    
    df = df.iloc[3:,:]
    df = df.loc[:20,:]
    df.drop(columns=['Unidade Geográfica', 
                  '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Anos Iniciais - EF', 'Anos Finais - EF', 'Total - EF',
                  '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 'Total - EM', 
                  'R_EF_Anos Iniciais', 'R_EF_Anos Finais',], inplace=True)
    
    return df
```

```python
query = ['Rural', 'Urbana']
urbana_rural = fail_dataset.query('Rede == "Total" and Localização in @query')
urbana_rural
```

```python
query = ['Privada', 'Particular', 'Publico', 'Pública']
privada_particular = fail_dataset.query('Localização == "Total" and Rede in @query')
privada_particular
```

```python
final_fail_dataset = pd.concat([urbana_rural, privada_particular])
final_fail_dataset
```

```python
def format_rede(value):
    if value == 'Publico':
        return 'Pública'
    elif value == 'Particular':
        return 'Privada'
    else:
        return value
    
final_fail_dataset.Rede = final_fail_dataset.Rede.apply(format_rede)
final_fail_dataset
```
