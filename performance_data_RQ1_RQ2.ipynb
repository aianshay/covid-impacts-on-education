```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""works for year > 2010"""

def create_performance_dataframe(year):
    df = pd.read_excel(f'data/tx_rend_brasil_regioes_ufs_{year}.xlsx', skiprows=5)
    
    #selecting columns
    df = df.iloc[:,:22]
    
    #naming columns
    df.columns = ['Ano', 'Unidade Geográfica', 'Localização',
                  'Rede', 'Total - EF', 'Anos Iniciais - EF',
                  'Anos Finais - EF', '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Total - EM', '1EM', '2EM', '3EM', '4EM', 'Não-Seriado']
    
    df = df.iloc[3:,:]
    df = df.loc[:20,:]
    df.drop(columns=['Unidade Geográfica', 'Anos Iniciais - EF', 'Anos Finais - EF'], inplace=True)
    
    return df
    
performance_2020 = create_performance_dataframe(2020)
performance_2020
```

```python
"""works for year <= 2010"""

def create_performance_dataframe_updated(year, skiprows=5):
    df = pd.read_excel(f'data/tx_rend_brasil_regioes_ufs_{year}.xlsx', skiprows=skiprows)
    
    #selecting columns
    df = df.iloc[:,:22]
    
    #naming columns
    df.columns = ['Ano', 'Unidade Geográfica', 'Rede',
                  'Localização', '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                  '8EF', '9EF', 'Anos Iniciais - EF', 'Anos Finais - EF', 'Total - EF',
                  '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 'Total - EM']
    
    df = df.iloc[3:,:]
    df = df.loc[:20,:]
    df.drop(columns=['Unidade Geográfica', 'Anos Iniciais - EF', 'Anos Finais - EF'], inplace=True)
    
    return df
```

```python
performance_2010 = create_performance_dataframe_updated(2010)
performance_2010
```

```python
performance_2009 = create_performance_dataframe_updated(2009, skiprows=6)
performance_2009
```

```python
performance_2008 = create_performance_dataframe_updated(2008)
performance_2008.columns = ['Ano', 'Localização', 'Rede',
                            '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                            '8EF', '9EF', 'Total - EF',
                            '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 'Total - EM']
performance_2008
```

```python
performance_2007 = create_performance_dataframe_updated(2007)
performance_2007.columns = ['Ano', 'Localização', 'Rede',
                            '1EF', '2EF', '3EF', '4EF', '5EF', '6EF', '7EF',
                            '8EF', '9EF', 'Total - EF',
                            '1EM', '2EM', '3EM', '4EM', 'Não-Seriado', 'Total - EM']
performance_2007
```

```python
perfs = [performance_2020, performance_2019, performance_2018, performance_2017, performance_2016,
         performance_2015, performance_2014, performance_2013, performance_2012, performance_2011,
         performance_2010, performance_2009, performance_2008, performance_2007]

perf_data = pd.concat(perfs)
perf_data.head()
```
