```python
import pandas as pd
import os
```

```python
df_2018_2010 = pd.DataFrame(columns=['#ANO', 'ESTADO', 'MUNICIPIO', 'ESFERA_GOVERNO', 
                                     'MODALIDADE_ENSINO', 'VALOR_TOTAL_ESCOLAS_PNAE'])

for file in os.listdir('data/pnae/'):
    if file.endswith('.csv'):
        df = pd.read_csv(f'data/pnae/{file}', encoding="ISO-8859-1", sep=';')
        df_2018_2010 = df_2018_2010.append(df)
        print(df.columns)
        
df_2018_2010.head()
```

```python
df_2018_2010['#ANO'].unique()
```

```python
df_2019 = pd.read_csv('data/pnae/2019/RecursosRepassados2019.csv')
df_2019.head()
```

```python
df_2019 = df_2019.drop(columns=['Co_recursos_repassados'])
df_2019.head()
```

```python
new_cols = {'Ano':'#ANO',
            'Estado':'ESTADO',
            'Municipio':'MUNICIPIO',
            'Esfera_governo':'ESFERA_GOVERNO',
            'Modalidade_ensino':'MODALIDADE_ENSINO',
            'Vl_total_escolas':'VALOR_TOTAL_ESCOLAS_PNAE'}
            
df_2019.rename(columns=new_cols, inplace=True)
df_2019.head()
```

```python
full_df = df_2018_2010.append(df_2019)
```

```python
full_df.head()
```

```python
full_df.info()
```

```python
full_df['#ANO'].unique()
```

```python
full_df['ESTADO'].unique()
```

```python
def format_number(value):
    value = str(value)
    return value.replace(',', '.')

full_df['VALOR_TOTAL_ESCOLAS_PNAE'] = full_df['VALOR_TOTAL_ESCOLAS_PNAE'].apply(format_number)
full_df['VALOR_TOTAL_ESCOLAS_PNAE'] = pd.to_numeric(full_df['VALOR_TOTAL_ESCOLAS_PNAE'])
```

```python
pnae_data = pd.DataFrame(full_df.groupby(['#ANO', 'ESTADO'])['VALOR_TOTAL_ESCOLAS_PNAE'].sum()).reset_index()

pnae_data.head()
```

```python
pnae_data = pnae_data.sort_values(['#ANO', 'ESTADO'])
```

```python
pnae_data['#ANO'].unique()
```

```python
pnae_data['ESTADO'].unique()
```

```python
new_cols = {"#ANO":"year",
            "ESTADO": "state",
            "VALOR_TOTAL_ESCOLAS_PNAE":"pnae_budget"}

pnae_data = pnae_data.rename(columns=new_cols)
```

```python
pnae_data.to_csv('data/yearly_state_pnae_budget.csv', index=False)
```
